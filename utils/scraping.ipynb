# Ancient Greek

## Lexicon

```python
from tqdm import tqdm
```

```python
list_all_word = []
for num in tqdm(range(0, 42021, 20)):
    source = urllib.request.urlopen(f'https://www.greek-language.gr/digitalResources/ancient_greek/tools/liddel-scott/search.html?start={num}&lq=').read()
    soup = bs.BeautifulSoup(source,'lxml')
    mydivs = soup.find_all("div", {"class": "ancient_greek"})
    words = str(mydivs).split('<dd>')
    for word in words:
        token = re.findall('"lemma">(.*?)</b>', word)
        print(token)
        list_all_word.append(token)
```

```python
import numpy as np
```

```python
# Remove empty List from List
# using list comprehension
list_all_word = [ele for ele in list_all_word if ele != []]
```

```python
np.savetxt("lexicon/lexicon_ancient_greek.csv", 
           list_all_word,
           delimiter =", ", 
           fmt ='% s')
```

```python
with open(r'lexicon/lexicon_ancient_greek.txt', 'w') as fp:
    for item in list_all_word:
        # write each item on a new line
        fp.write("%s\n" % item)
    print('Done')
```

```python
# http://users.uoa.gr/~nektar/history/tributes/ancient_authors/index.htm

for a in soup.find_all('a', href=True):
    ######################################
    url = a['href']
    name_text = url.split('.')[0].replace("/", "_")
    print("Found the URL:", )
    print("Name:",a.string)    
    
    #######################################
    text_url = basic_url + url
    print(text_url)
    text_src = urllib.request.urlopen(text_url).read()
    text_soup = bs.BeautifulSoup(text_src,'lxml')
    ANCIENT_TEXT = str(text_soup.find_all('p'))

    # remove HTML tags from a string in Python
    regex = re.compile(r'<[^>]+>')
    def remove_html(string):
        return regex.sub('', string)
    ANCIENT_TEXT = remove_html(ANCIENT_TEXT)

    with open( 'ancient_greek/'+name_text+'.txt', 'w') as f:
         f.write('TITLE: ' + a.string + '\n')
         f.write(ANCIENT_TEXT)
    ####################################### 
    if a.string == 'ΒΑΤΡΑΧΟΜΥΟΜΑΧΙΑ':
        break    
```

```python
import time
buz_url = 'https://byzantium.gr/keimena/'
def retrieve_text(url):
    time.sleep(4)
    wd.get(url)
    soup = bs.BeautifulSoup(wd.page_source)
    # print(soup)
    firstdivs = str(soup.find_all("td", {"class": "bla1 blax2"}))

    # print(mydivs)

    results_all_books = re.findall('href="(.*)">', str(firstdivs))  # Find string between two substrings [duplicate]
    results_all_books = list(dict.fromkeys(results_all_books))  # Delete Duplicate
    results_all_books = [x for x in results_all_books if not x.startswith('https://')]  #remove all list elements starting with http
    print(results_all_books)

    all_books = []
    if results_all_books!=[]: # sub books if exists
        for book in results_all_books:
            extra_text = extra_retrieve_text(buz_url+book)
            all_books.append(extra_text)
    

    text = re.sub('href="(.*)</div>', "", str(firstdivs))
    text = re.sub('<a(.*)</a>', "", text)
    text = re.sub('<i(.*)</i>', "", text)
    text = re.sub('<span(.*)</span>', "", text)
    text = re.sub('<div(.*)</div>', "", text)

    # remove HTML tags from a string in Python
    regex = re.compile(r'<[^>]+>')
    def remove_html(string):
        return regex.sub('', string)
    new_text = remove_html(text)
    
    new_text = "".join([s for s in new_text.splitlines(True) if s.strip("\r\n")])

    all_books.append(new_text)

    # pprint(all_books)

    return all_books
```
